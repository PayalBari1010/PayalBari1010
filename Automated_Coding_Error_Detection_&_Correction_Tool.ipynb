```python
import os
import pandas as pd

# Function to read code files from a directory
def load_code_files(directory_path):
    code_files = []
    for folder, _, files in os.walk(directory_path):
        for file_name in files:
            if file_name.endswith('.py') or file_name.endswith('.cpp') or file_name.endswith('.c'):
                file_path = os.path.join(folder, file_name)
                with open(file_path, 'r', encoding="utf-8") as file:
                    code = file.read()
                    code_files.append({'file_name': file_name, 'code': code})
    return pd.DataFrame(code_files)

# Load dataset (assuming data is in a 'data' folder within 'mini_project_codenet')
data_dir = '/content/data'
code_data = load_code_files(data_dir)

# Preview the first few code snippets
print(code_data.head())
print(code_data.tail())
```

```python
import os
import pandas as pd

# Function to load code files and associate them with metadata
def load_code_and_labels(data_directory, metadata_directory):
    code_data = []

    # Iterate through metadata files to get labels
    for metadata_file in os.listdir(metadata_directory):
        if metadata_file.endswith('.csv'):
            metadata_path = os.path.join(metadata_directory, metadata_file)
            metadata_df = pd.read_csv(metadata_path)

            # Extract problem ID from metadata filename
            problem_id = metadata_file.split('.')[0]
            problem_folder_path = os.path.join(data_directory, problem_id)

            if os.path.exists(problem_folder_path):
                # Combine submission_id and filename_ext to construct filenames
                metadata_df['filename'] = metadata_df['submission_id'].astype(str) + '.' + metadata_df['filename_ext']

                # Traverse language-based subfolders
                for language_folder in os.listdir(problem_folder_path):
                    lang_path = os.path.join(problem_folder_path, language_folder)
                    if os.path.isdir(lang_path):
                        # Process code files within the language folder
                        for code_file in os.listdir(lang_path):
                            if code_file in metadata_df['filename'].values:
                                file_path = os.path.join(lang_path, code_file)
                                with open(file_path, 'r', encoding='utf-8') as file:
                                    code = file.read()

                                # Find the status/label from metadata
                                status_row = metadata_df[metadata_df['filename'] == code_file]
                                if not status_row.empty:
                                    status = status_row['status'].values[0]
                                    code_data.append({'file_name': code_file, 'code': code, 'label': status})
                            else:
                                print(f"Code file {code_file} not found in metadata for problem ID {problem_id}.")

    return pd.DataFrame(code_data)

# Paths to data and metadata folders
data_directory = '/content/data'
metadata_directory = '/content/metadata'

# Load dataset
dataset = load_code_and_labels(data_directory, metadata_directory)

# Preview the dataset
print(dataset.head())
```

```python
# Map statuses to binary or categorical labels
def map_labels(status):
    if status == 'Accepted':
        return 'No Error'
    elif status in ['Runtime Error', 'Wrong Answer', 'WA: Presentation Error', 'Time Limit Exceeded', 'Compile Error', 'Memory Limit Exceeded']:
        return 'Error'
    else:
        return 'Unknown'

# Apply label mapping
dataset['mapped_label'] = dataset['label'].apply(map_labels)

# Display label distribution
print(dataset['mapped_label'].value_counts())
```

```python
dataset = dataset[dataset['mapped_label'] != 'Unknown']
```

Code Tokenization and Cleaning:

You may want to remove comments and unnecessary whitespace from the code files.

Use regular expressions (regex) to extract meaningful *content*

```python
import re
def clean_code(code):
    # Remove comments
    code = re.sub(r'//.*|/\*[\s\S]*?\*/', '', code)
    # Remove excess whitespace
    return re.sub(r'\s+', ' ', code).strip()

dataset['cleaned_code'] = dataset['code'].apply(clean_code)
dataset
```

 Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)  # Adjust max_features as needed
X = vectorizer.fit_transform(dataset['cleaned_code']).toarray()
print(X)
```

 Split Data for Training and Testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, dataset['mapped_label'], test_size=0.2, random_state=42
)
```

Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
grid.fit(X_train, y_train)
print(grid.best_params_)
```

```python
import gradio as gr
import autopep8
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load GPT-Neo (Open Source Model from HuggingFace)
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-1.3B")
model = AutoModelForCausalLM.from_pretrained("EleutherAI/gpt-neo-1.3B").to("cuda" if torch.cuda.is_available() else "cpu")

# Error explanations dictionary
error_explanations = {
    'SyntaxError': "There is a syntax issue. Make sure to use the correct symbols, colons, and indentation.",
    'NameError': "You're using a variable or function that hasn't been defined. Double-check your variable names.",
    'TypeError': "A function or operation is applied to an incompatible type. Check your data types.",
    'ZeroDivisionError': "The code tries to divide by zero. Add checks to avoid dividing by zero.",
    'IndentationError': "There is an indentation issue. Ensure consistent use of spaces or tabs.",
    'No Error': "Your code has no errors!"
}

def predict_code_error(user_code):
    """Detect if the input code has an error and return the error type."""
    try:
        exec(user_code)  # Try to execute the code
        return "No Error"
    except Exception as e:
        return type(e).__name__  # Return the error type

def get_error_explanation(error_type):
    """Provide detailed explanations for the detected error."""
    return error_explanations.get(error_type, "No specific explanation available.")

def auto_fix_code(user_code):
    """Automatically fix syntax and logical errors using autopep8 and GPT-Neo."""
    try:
        # Auto-fix basic syntax using autopep8
        syntax_fixed_code = autopep8.fix_code(user_code)

        # Use GPT-Neo for advanced code fixes (logic errors)
        input_text = f"Fix the following code:\n{syntax_fixed_code}"
        inputs = tokenizer(input_text, return_tensors="pt").to(model.device)
        outputs = model.generate(**inputs, max_new_tokens=150, temperature=0.5)

        fixed_code = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return fixed_code.strip()

    except Exception as e:
        return f"Error during auto-fix: {str(e)}"

def detect_and_fix(user_code):
    """Detect errors, explain them, and attempt to fix the code."""
    error_type = predict_code_error(user_code)
    explanation = get_error_explanation(error_type)

    if error_type == "No Error":
        return f"Congrats! No errors detected.\n\nExplanation: {explanation}"
    else:
        fixed_code = auto_fix_code(user_code)
        return f"Error Type: {error_type}\nExplanation: {explanation}\n\nSuggested Fix:\n{fixed_code}"

# Gradio Interface with two tabs: Detect & Fix, and Manual Fix
with gr.Blocks() as app:
    gr.Markdown("# Code Error Detector & Auto-Fixer")

    with gr.Tabs():
        with gr.TabItem("Detect & Fix"):
            user_code = gr.Textbox(label="Enter your code snippet", lines=10, placeholder="Type your code here...")
            detect_button = gr.Button("Detect Error")
            result = gr.Textbox(label="Result", lines=10, interactive=False)

            def on_detect_click(code):
                return detect_and_fix(code)

            detect_button.click(on_detect_click, inputs=user_code, outputs=result)

        with gr.TabItem("Manual Fix"):
            manual_input = gr.Textbox(label="Enter your code snippet", lines=10, placeholder="Type your code here...")
            fix_button = gr.Button("Fix It!")
            fixed_result = gr.Textbox(label="Fixed Code", lines=10)

            def on_fix_click(code):
                return f"Your code:\n{code}\n\nTry fixing the error manually."

            fix_button.click(on_fix_click, inputs=manual_input, outputs=fixed_result)

# Launch the Gradio app
app.launch()
```
